1. Line plot

```python
a = np.arange(1,11)
b = np.arange(11,21)
plt.plot(a,b)
```

```python
a = np.arange(1,11)
b = np.arange(11,21)
plt.plot(a,b)
plt.title("Line plot")
plt.xlabel("values ranging from 1 to 11")
plt.ylabel("values ranging from 11 to 21")
```

```python
a = np.arange(1,11)
b = np.arange(11,21)
plt.plot(a,b,color="red",linewidth = 5)
plt.title("Line plot")
plt.xlabel("values ranging from 1 to 11")
plt.ylabel("values ranging from 11 to 21")
```

```python
days = list(range(1,8))
temp = [34,35,40,41,42,34,44]
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Weather report")
plt.plot(days,temp,color="green",linewidth=5)
```

```python
days = list(range(1,8))
temp = [34,35,40,41,42,34,44]
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Weather report")
plt.plot(days,temp,color="green",linewidth=5)
plt.xlim([1,10])
plt.ylim([34,46])
```

```python
x = np.arange(10)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,x,label="x values")
plt.plot(x,2*x,label="2*x values")
plt.plot(x,3*x,label="3*x values")
plt.plot(x,4*x,label="4*x values")
plt.legend(loc="upper left")
plt.show()
```

```python
x = np.arange(10)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,x,label="x values")
plt.plot(x,2*x,label="2*x values")
plt.plot(x,3*x,label="3*x values")
plt.plot(x,4*x,label="4*x values")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
x = np.arange(10)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,x,"r--",x,2*x,"b*",x,3*x,"y^",x,4*x,"g+")
#plt.legend(loc="upper left")
plt.grid()
plt.show()
```
